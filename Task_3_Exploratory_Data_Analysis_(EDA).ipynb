```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/miirshe/USA-Housing-Analysis-and-Prediction-Price/main/USA_Housing.csv"
USAhousing = pd.read_csv(url)
```

```python
USAhousing.head()
USAhousing.info()
USAhousing.shape
```

```python
#Check missing values
USAhousing.isnull().sum()
```

```python
USAhousing.describe()
```

```python
USAhousing.describe(include="object")
```

```python
USAhousing.select_dtypes(include=['number']).median()
```

```python
USAhousing.hist(figsize=(12,8))
plt.show()
```

```python
sns.histplot(USAhousing["Price"], kde=True)
plt.title("Price Distribution")
plt.show()
```

```python
sns.scatterplot(x="Avg. Area House Age", y="Price", data=USAhousing)
plt.title("Avg. Area House Age vs Price")
plt.show()
```

```python
USAhousing.columns
```

```python
sns.boxplot(x="Avg. Area House Age", y="Price", data=USAhousing)
plt.title("Price by Avg. Area House Age")
plt.show()
```
